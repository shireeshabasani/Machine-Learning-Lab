## print other metrics

```python
 import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load dataset
df = pd.read_csv('/home/user/Downloads/Iris.csv')

# Prepare features and target
x = df.drop(columns=['Id', 'Species'])
y = df['Species']

# Split into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

# Build and train model
model = SVC(C=1)
model.fit(x_train, y_train)

# Predict on test set
y_pred = model.predict(x_test)

# Print evaluation metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision (macro avg):", precision_score(y_test, y_pred, average='macro'))
print("Recall (macro avg):", recall_score(y_test, y_pred, average='macro'))
print("F1-score (macro avg):", f1_score(y_test, y_pred, average='macro'))
```

## print classification report and accuracy 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

df = pd.read_csv('/home/user/Downloads/Iris.csv')
x = df.drop(columns=['Id', 'Species'])
y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

model = SVC(C=1)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## k fold cross validation

```python
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold
from sklearn.svm import SVC

df = pd.read_csv('/home/user/Downloads/Iris.csv')
x = df.drop(columns=['Id', 'Species'])
y = df['Species']
model = SVC(C=1)
kf = KFold(n_splits=5)
scores = cross_val_score(model, x, y, cv=kf)
print("Cross-validation scores:", scores)
```
